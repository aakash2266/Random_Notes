```python
import keras
import tensorflow as tf
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.imdb.load_data(num_words=1000)
```

```python
X_train.shape
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(X_train, maxlen=500)
X_test = pad_sequences(X_test, maxlen=500)
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding

model = Sequential([
    Embedding(10000,16),
    SimpleRNN(32,return_sequences=True),
    SimpleRNN(32,activation='tanh'),
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
loss,acc = model.evaluate(X_test,y_test)
print(acc)
```

```python
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding,LSTM
model1 = Sequential([
    Embedding(10000,16),
    LSTM(32,return_sequences=True),
    LSTM(32,activation='tanh'),
    Dense(1,activation='sigmoid')
])
```

```python
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model1.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
loss1,acc1 = model1.evaluate(X_test,y_test)
print(acc1)
```
